PART-2:VIZUALIZATION OF DATA USING MATPLOTLIB,PYPLOTS,PACKAGES
1. GENERATE A DATASET FOR HEALTH CARE USING THE FOLLOWING


```python
import matplotlib.pyplot as plt
```

```python
people=["charvika","chodhitha","shivani","sharanya"]
age=[18,18,19,19]
height=[150,154,147,140]
weight=[40,42,45,43]
```

```python
plt.scatter(weight,height)
plt.title("relationship between weight and height of the patients")
plt.ylabel("height in metres")
plt.xlabel("weight of the patients in kgs")
plt.show()
```

BAR PLOT

```python
plt.figure(figsize=(15,5))
plt.title("weight of people in kgs",fontsize=16,fontstyle='oblique')
plt.bar(x=people,height=weight,width=0.5,color='green',alpha=0.8)
plt.xlabel("people",fontsize=14)
plt.xticks(fontsize=12,rotation=20)
plt.yticks(fontsize=12)
plt.ylabel("weight in kgs",fontsize=13)
plt.show()
```

HISTOGRAM

```python
plt.figure(figsize=(8,7))
plt.hist(weight,color='blue',edgecolor='k',alpha=0.70,bins=5)
plt.title("histogram of patient's wight",fontsize=16)
plt.xlabel("weights in kgs",fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
